```python
import os, glob

def load_data_for_count(file_path, label_list):
    count = 0
    data_output = []
    
    for file in glob.glob(file_path):
        file_name = os.path.basename(file)
        file_name_components = file_name.split("_")
        for component in file_name_components:
            if component in label_list:
                data_output.append(component)
    return np.array(data_output)
```

## Dataset Organization 

```python
#function to take the lists of categories to organize the directories
import os

def make_category_directories(category_list, rootpath):
    for category in category_list:
        os.makedirs(str(rootpath) + '/' + str(category))
```

```python
os.makedirs('angle_data')
os.makedirs('angle_data/train')
os.makedirs('angle_data/test')
os.makedirs('angle_data/val')
```

```python
os.makedirs('level_data')
os.makedirs('level_data/train')
os.makedirs('level_data/test')
os.makedirs('level_data/val')
```

```python
shot_levels = ["aerial", "eye", "ground-knee", "hip", "shoulder"]
shot_angles = ["dutch", "high", "low", "neutral", "overhead"]
```

```python
angle_dir = 'angle_data'
level_dir = 'level_data'
```

```python
# Putting folders of the classes in each Angle Set
make_category_directories(shot_angles, 'angle_data/train') 
make_category_directories(shot_angles, 'angle_data/test') 
make_category_directories(shot_angles, 'angle_data/val') 
```

```python
# Putting Folders for the classes in each set for levels
make_category_directories(shot_levels, 'level_data/train') 
make_category_directories(shot_levels, 'level_data/test') 
make_category_directories(shot_levels, 'level_data/val') 
```

```python
# variables for test/train/val ratios (in decimal)
train_rat = 0.7
test_rat = 0.15
val_rat = 0.15
```

```python
#Functions to get random selected samples from the raw dataset and assign to directory 
import random, os, glob, shutil

# Function to get random sample from dataset given a dataset
def randomized_sample_list(raw_path, sample_size):
    path_list = []
    file_sample = random.sample(list(os.scandir(raw_path)), sample_size)
    for file in file_sample:
        path_list.append(file.path)
    return path_list

# Function for Splitting list data into allocated sections for dividing train/test/val
def split_sample(sample_list, train_r, test_r, val_r):
    sample_num = len(sample_list)
    train,test,val = [],[],[]
    # determine how many paths to go into each section
    train_num = int(train_r*sample_num)
    test_num = int(test_r*sample_num)
    val_num = int(val_r*sample_num)
    for i in range(sample_num):
        if i < train_num:
            train.append(sample_list[i])
        elif i > train_num and i < (train_num + test_num):
            test.append(sample_list[i])
        else:
            val.append(sample_list[i])
    return train, test, val
    
# Function for moving those sections into the directories
def copy_files(path_list, destin_directory):
    for path in path_list:
        shutil.copy(path, destin_directory)
    print(str(len(path_list)) +" files copied to " + destin_directory)

# Function that Combines all previous functions to split the data
def split_data(raw_path, destination_path, labels, sample_size):
    for folder in glob.glob(raw_path):
        folder_name = os.path.basename(folder)
        folder_path = raw_path.replace("*", "") + folder_name
        folder_dir_list = list(os.scandir(folder_path))
        if  len(folder_dir_list) >= sample_size:
            sample = randomized_sample_list(folder_path, sample_size)
            train, test, val = split_sample(sample, train_rat, test_rat, val_rat)
            train_path = destination_path + '/train'+ '/' + str(folder_name)
            test_path = destination_path + '/test'+ '/' + str(folder_name)
            val_path = destination_path + '/val'+ '/' + str(folder_name)
            if folder_name in labels:
                copy_files(train, train_path)
                copy_files(test, test_path)
                copy_files(val, val_path) 

# Function that dumps all data from raw to the root of the destination
def dump_data(raw_path, destination_path, labels, subfolder):
    
    for folder in glob.glob(raw_path):
        folder_name = os.path.basename(folder)
        if folder_name in labels: 
            folder_path = raw_path.replace("*", str(folder_name))
            new_destin = destination_path + '/' + subfolder + '/' + str(folder_name)
            path_list = []
            for file in os.listdir(folder_path):
                path_list.append(folder_path + '/' + file)
            copy_files(path_list, new_destin)
```

```python
# Testing Functions 
test_list = randomized_sample_list('Raw Datasets/CineScale2/angle/dutch', 20)
test_x, test_y, test_z = split_sample(test_list, train_rat, test_rat, val_rat)
print(test_x)
print(test_y)
print(test_z)
copy_files(test_y, 'test dir')
```

```python
# Angle Data Importing for Use to Train
split_data('Raw Datasets/CineScale2/angle/*', angle_dir, shot_angles, 471)
dump_data('Raw Datasets/Self-created/angle/*', angle_dir, shot_angles,'train')
```
